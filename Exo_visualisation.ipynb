Nombre d’host et nombre de review par quartier. 

```python
 df_bordeaux = pd.read_csv('/home/apprenant/Documents/Data/bordeaux.csv')
```

```python
df_bordeaux_neighbourhood = df_bordeaux[["host_id","neighbourhood_cleansed"]].groupby("neighbourhood_cleansed").agg({"host_id":"count"})
df_bordeaux_neighbourhood
```

```python
df_bordeaux_neighbourhood.plot.bar(figsize=(20,5))

plt.title("Nombre d'host par quartier")
plt.xlabel("Quartiers")
plt.ylabel("Nombre d'host")

```

```python
df_bordeaux_nb_host = df_bordeaux[["number_of_reviews","neighbourhood_cleansed"]].groupby("neighbourhood_cleansed").agg({"number_of_reviews":"sum"})
df_bordeaux_nb_host
```

```python
df_bordeaux_nb_host.plot.bar(figsize=(20,5), color="orange")

plt.title("Nombre de reviews par quartier")
plt.xlabel("Quartiers")
plt.ylabel("Nombre de reviews")
```

Pourcentage des hosts ayant une vérification par type.

```python
df_verefication = df_bordeaux[['host_verifications',"host_id"]]
totale_verification = df_verefication.shape[0]

number_phone = df_verefication[df_verefication['host_verifications'].str.contains('phone')]['host_verifications'].count()
porcent_phone =(number_phone/totale_verification)*100

number_email = df_verefication[df_verefication['host_verifications'].str.contains('email')]['host_verifications'].count()
porcent_email = (number_email/totale_verification)*100

number_email_work = df_verefication[df_verefication['host_verifications'].str.contains('work_email')]['host_verifications'].count()
porcent_email_work = (number_email_work/totale_verification)*100

plt.bar(['téléphone', 'email', 'email professionnel'], [porcent_phone, porcent_email, porcent_email_work], color='green')

plt.title('Pourcentage des vérifications selon le moyen utilisé')
plt.xlabel('Moyen utilisé')
plt.ylabel('Pourcentage des vérifications')
```
